## CHAINING OPERATIONS

#### 1. Filter for respondents over 40 years old
#### 2. Select only the columns for degree, income and sex
#### 3. Create a new column "income category" that labels income as "high" if over 50000 otherwise "low"
#### 4. Sort the results by income in descending order (Highest incomes first)

```python
gss_processed=(
    gss.query("age>40")
    .loc[:,["degree", "income", "sex"]]
    .assign(income_category=lambda x:np.where(x["income"]>50000,"High","Low")).sort_values(by="income",ascending=False))
gss_processed
```
